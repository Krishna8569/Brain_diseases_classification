# Image classification using PCA

We will apply PCA to the images and calculate the difference between the two largest eigenvalues of the covariance matrix and scatter plot them for each of the figure types to see whether they show any pattern for classification
***

```python
## import required modules
from sklearn.decomposition import PCA, IncrementalPCA
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

### 1.*Non Dementia*
***

```python
%%timeit


## 
n_fig = 640
NonDem_lambda1 = []
NonDem_lambda2 = []
for fig in range(n_fig):
    image_bw = cv2.imread(f"Alzheimer_s_Dataset/train/NonDemented/nonDem{fig}.jpg",cv2.IMREAD_GRAYSCALE)
    pca = PCA()
    pca.fit(image_bw)
    lambda_val = pca.explained_variance_[0:2]
    ## obtaining the two largest eigenvalues
    NonDem_lambda1.append(lambda_val[0])
    NonDem_lambda2.append(lambda_val[1])
    
NonDem_lambda1 = np.array(NonDem_lambda1)
NonDem_lambda2 = np.array(NonDem_lambda2)
## plotting the data
plt.scatter(NonDem_lambda1,NonDem_lambda2,c='b',label='NonDem')
plt.legend()
plt.show()
## plotting the asymmetry
asymmetry_no = NonDem_lambda1 - NonDem_lambda2
plt.plot(asymmetry_no)
plt.show()
```
